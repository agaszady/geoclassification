pobranie i filtrowanie miejscowosci 

```python
miejscowosci_prng = gpd.read_file('PRNG_MIEJSCOWOSCI_SHP/PRNG_MIEJSCOWOSCI_SHP.shp')
```

```python
lubelskie_prng = miejscowosci_prng[miejscowosci_prng['wojewodz'] == 'lubelskie']
```

zamiana kolumn z id simc, idMscNd na pola numeryczne, wyrzucenie przypadków bez simc

```python
lubelskie_prng['idZewnetrz'] = pd.to_numeric(lubelskie_prng['idZewnetrz'], errors='coerce').astype('Int64')
lubelskie_prng['idMscNd'] = pd.to_numeric(lubelskie_prng['idMscNd'], errors='coerce').astype('Int64')
lubelskie_prng = lubelskie_prng[['idZewnetrz', 'rodzaj', 'nazwaGlown', 'idMscNd', 'nazwaMscNd', 'geometry']]
lubelskie_prng.rename(columns={'idZewnetrz': 'SIMC_id'}, inplace=True)
lubelskie_prng.dropna(subset=['SIMC_id'], inplace=True)
```

wyrzucenie miast z analizy

```python
# df_nazwy = lubelskie_prng[['SIMC_id', 'rodzaj', 'nazwaGlown']]
condition = lubelskie_prng['rodzaj'].isin(['miasto', 'część miasta', 'osiedle'])
lubelskie_prng = lubelskie_prng[~condition]
lubelskie_prng.drop_duplicates(inplace=True)
```

```python
lubelskie_prng
```

```python
lubelskie_prng[['rodzaj', 'SIMC_id']].groupby(by=['rodzaj']).count()
```

dodanie kolumny czy_samodzielna

```python
lubelskie_prng['czy_samodzielna'] = [True if pd.isna(id_msc_nd) else False for id_msc_nd in lubelskie_prng['idMscNd']]
lubelskie_prng
```
